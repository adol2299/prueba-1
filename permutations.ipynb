```
import itertools

## Method to divide Digits ##
def divide(c):
 #Declaration and filling of matrix
 perm1=[]
 perm2=[]
 for i in str(c):
   perm1.append(i)

  #Permutate the array
 permut = itertools.permutations(perm1)
 x=1
 for y in permut:
   perm2.append([x,y])
   x+=1

 #Print the permutations
 print("5-Permutaciones del Conjunto [1,2,3,4,5] \n")
 for r in perm2:
   print(r)
 imprimir(perm2)  

## Method composition of permutations
def comp(mat1, mat2):
 mat3 = [mat2[int(mat1[0])-1], mat2[int(mat1[1])-1], mat2[int(mat1[2])-1], mat2[int(mat1[3])-1], mat2[int(mat1[4])-1]]
 return mat3

## Method to print and organize the compositions
def imprimir(matriz):
  size = len(matriz) + 1
  print (size)
  matr = []
  for i in range(0, size):
    mil = []
    for j in range(0, size):
      if(i == 0 and j == 0):
        print("\n\n\n                         ", end=" ")
      elif(i == 0 and j != 0):
        print (matriz[j-1][1], end=" ")
      elif(j==0 and i != 0):
        print (matriz[i-1][1], end=" ")
      elif(i != 0 and j != 0):
        a=(comp(matriz[i-1][1], matriz[j-1][1]))
        mil.append(a)
        print (a, end=" ")
    matr.append(mil)
    print(" ")


divide(12345)


```
